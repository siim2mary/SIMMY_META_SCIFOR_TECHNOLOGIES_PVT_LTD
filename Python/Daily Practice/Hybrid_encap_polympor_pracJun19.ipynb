Hybrid Inheritance in Python
Hybrid inheritance is a combination of two or more types of inheritance. Here, we will create an example incorporating encapsulation and polymorphism.

Create a class structure representing a basic organization hierarchy:

Employee: Inherits from Person and has additional attributes like employee ID and department.
Manager: Inherits from Employee and has a list of employees they manage.
Developer and Designer: Both inherit from Employee but have specific roles.
TechLead: Inherits from both Developer and Manager.
Implement encapsulation to protect the data members.

Use polymorphism to create a common interface for all employees to display their details.

```python
class Orga:
    def __init__(self, name, age):
        self.__name = name  # Encapsulated data member
        self.__age = age    # Encapsulated data member

    def get_name(self):
        return self.__name

    def get_age(self):
        return self.__age

    def details(self):
        print(f"The details of person in Organisation: {self.get_name()} and {self.get_age()}")

class Person(Orga):
    def __init__(self, name, age):
        super().__init__(name, age)

    def details(self):
        print(f'The person in Organisation {self.get_name()} and his age is {self.get_age()}')

class Employee(Person):
    def __init__(self, name, age, empID, dept):
        super().__init__(self,dept)
        Person.__init__(self,name , age)
        self.name = name
        self.age = age
        self.empID = empID
        self.dept = dept

    def per_det(self):
        print(f'The Organisational details of employee are {self.empID} and {self.dept}')

    def details(self):
        print(f'The Employee details {self.get_name()} and {self.get_age()} with empID {self.empID} in {self.dept}')

class Manager(Employee):
    def __init__(self, name, age, empID, dept, emplist):
        super().__init__(name, age, empID, dept)
        self.emplist = emplist

    def details(self):
        print(f'The employee under Manager are: {self.get_name()} and {self.get_age()} with empID {self.empID} in {self.dept}')
        print(f'The employee list under Manager is {self.emplist}')

class Developer(Employee):
    def __init__(self, name, age, empID, dept, des):
        super().__init__(self,name,age,empID, des)
        self.des = des

    def role(self):
        print(f'The designation of employee in developer: {self.des}')

    def details(self):
        super().details()
        print(f'The designation of employee in developer: {self.des}')

class Desig(Employee):
    def __init__(self, name, age, empID, dept, desig):
        super().__init__(name, age, empID, dept)
        self.desig = desig

    def details(self):
        super().details()
        print(f'The Designer works as Employee: {self.desig}')

class Techlead(Developer, Manager):
    def __init__(self, name, age, empID, dept, des, emplist):
        Developer.__init__(self, name, age, empID, dept, des)
        Manager.__init__(self, name, age, empID, dept, emplist)

    def details(self):
        Developer.details(self)
        print(f'Leads Team: {self.emplist}')

# Create instances
per = Person("Han", 23)
emp = Employee("Leia", 30, 1234, "Physics dept")
man = Manager("Luke", 40, 5678, "Management", ["Alice", "Bob"])
tech = Techlead("Prithu", 35, 9101, "Tech", "Lead Developer", ["Alice", "Bob"])

# Display details
print("Person Details:")
per.details()

print("\nEmployee Details:")
emp.details()

print("\nManager Details:")
man.details()

print("\nTechLead Details:")
tech.details()
```

```python
class Person:
    def __init__(self, name, age):
        self.__name = name  # Encapsulated data member
        self.__age = age    # Encapsulated data member

    def get_name(self):
        return self.__name

    def get_age(self):
        return self.__age

class Employee(Person):
    def __init__(self, name, age, emp_id, department):
        super().__init__(name, age)
        self.__emp_id = emp_id  # Encapsulated data member
        self.__department = department  # Encapsulated data member

    def get_emp_id(self):
        return self.__emp_id

    def get_department(self):
        return self.__department

    def display_details(self):
        print(f"Name: {self.get_name()}, Age: {self.get_age()}, Employee ID: {self.get_emp_id()}, Department: {self.get_department()}")

class Manager(Employee):
    def __init__(self, name, age, emp_id, department, employees=None):
        super().__init__(name, age, emp_id, department)
        if employees is None:
            employees = []
        self.__employees = employees  # Encapsulated data member

    def add_employee(self, employee):
        self.__employees.append(employee)

    def display_details(self):
        super().display_details()
        print(f"Manages: {[emp.get_name() for emp in self.__employees]}")

class Developer(Employee):
    def __init__(self, name, age, emp_id, department, programming_language):
        super().__init__(name, age, emp_id, department)
        self.__programming_language = programming_language  # Encapsulated data member

    def display_details(self):
        super().display_details()
        print(f"Programming Language: {self.__programming_language}")

class Designer(Employee):
    def __init__(self, name, age, emp_id, department, design_tool):
        super().__init__(name, age, emp_id, department)
        self.__design_tool = design_tool  # Encapsulated data member

    def display_details(self):
        super().display_details()
        print(f"Design Tool: {self.__design_tool}")

class TechLead(Developer, Manager):
    def __init__(self, name, age, emp_id, department, programming_language, employees=None):
        Developer.__init__(self, name, age, emp_id, department, programming_language)
        Manager.__init__(self, name, age, emp_id, department, employees)

    def display_details(self):
        Developer.display_details(self)
        print(f"Leads Team: {[emp.get_name() for emp in self._Manager__employees]}")

# Creating instances
dev1 = Developer("Alice", 30, "D123", "Development", "Python")
designer1 = Designer("Bob", 25, "D456", "Design", "Photoshop")
manager1 = Manager("Charlie", 40, "M789", "Management", [dev1, designer1])
techlead1 = TechLead("Eve", 35, "T101", "Tech Lead", "JavaScript", [dev1, designer1])

# Displaying details
print("Developer Details:")
dev1.display_details()

print("\nDesigner Details:")
designer1.display_details()

print("\nManager Details:")
manager1.display_details()

print("\nTechLead Details:")
techlead1.display_details()
```

```python
# Base Class
class Orga:
    def __init__(self, name, age):
        self.__name = name
        self.__age = age

    def get_name(self):
        return self.__name

    def get_age(self):
        return self.__age

    def details(self):
        print(f"The details of person in Organisation: {self.get_name()} and {self.get_age()}")

# Single Inheritance
class Person(Orga):
    def __init__(self, name, age):
        super().__init__(name, age)

    def details(self):
        super().details()
        print(f'The person in Organisation {self.get_name()} and his age is {self.get_age()}')

# Multilevel Inheritance
class Employee(Person):
    def __init__(self, name, age, empID, dept):
        super().__init__(name, age)
        self.empID = empID
        self.dept = dept

    def per_det(self):
        super().details()
        print(f'The Organisational details of employee are {self.empID} and {self.dept}')

    def details(self):
        super().details()
        print(f'The Employee details {self.get_name()} and {self.get_age()} with empID {self.per_det()}{self.empID} in {self.dept}')

# Another Base Class
class Manager:
    def __init__(self, name, age, empID, dept, emplist):
        self.name = name
        self.age = age
        self.empID = empID
        self.dept = dept
        self.emplist = emplist

    def manage(self):
        print(f'The Manager {self.name} with empID {self.empID} manages the following employees: {self.emplist}')

# Multiple Inheritance
class Developer(Employee):
    def __init__(self, name, age, empID, dept, des):
        super().__init__(name, age, empID, dept)
        self.__Des = des

    def role(self):
        super().details()
        Employee.per_det(self)
        print(f'The designation of employee in developer: {self.__Des}')

# Hybrid Inheritance
class Techlead(Developer, Manager):
    def __init__(self, name, age, empID, dept, des, emplist):
        Developer.__init__(self, name, age, empID, dept, des)
        Manager.__init__(self, name, age, empID, dept, emplist)

    def details(self):
        Developer.details(self)
        Manager.manage(self)

# Create Instances
person = Person("Han", 23)
employee = Employee("Leia", 30, 1234, "Physics dept")
developer = Developer("Luke", 35, 5678, "Development", "Senior Developer")
techlead = Techlead("Prithu", 40, 9101, "Tech", "Lead Developer", ["Alice", "Bob", "Charlie"])

# Display details
print("Person Details:")
person.details()

print("\nEmployee Details:")
employee.details()

print("\nDeveloper Details:")
developer.details()
developer.role()

print("\nTechLead Details:")
techlead.details()
```
