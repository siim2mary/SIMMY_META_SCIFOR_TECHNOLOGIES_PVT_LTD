```python
# Base Class
class Employee:
    def __init__(self, name, id, salary):
        self.name = name
        self.id = id
        self.salary = salary

    def display_details(self):
        print(f"Name: {self.name}")
        print(f"ID: {self.id}")
        print(f"Salary: ${self.salary}")

# Intermediate Classes
class FullTimeEmployee(Employee):
    def __init__(self, name, id, salary, benefits):  #always use self with def __init__
        super().__init__(name, id, salary) #Here super().__init__used since ist time taking dont use self
        self.benefits = benefits

    def display_details(self):
        super().display_details()  #__init__not used since want to display as in previous
        print(f"Benefits: {self.benefits}")

class PartTimeEmployee(Employee):
    def __init__(self, name, id, salary, hours_per_week):
        super().__init__(name, id, salary)
        self.hours_per_week = hours_per_week

    def display_details(self):
        super().display_details()
        print(f"Hours per week: {self.hours_per_week}")

# Derived Classes
class Manager(FullTimeEmployee):
    def __init__(self, name, id, salary, benefits, team_size):
        super().__init__(name, id, salary, benefits)
        self.team_size = team_size

    def display_details(self):
        super().display_details()
        print(f"Team Size: {self.team_size}")

class Developer(FullTimeEmployee):
    def __init__(self, name, id, salary, benefits, programming_languages):
        super().__init__(name, id, salary, benefits)
        self.programming_languages = programming_languages

    def display_details(self):
        super().display_details()
        print(f"Programming Languages: {', '.join(self.programming_languages)}")

class Intern(PartTimeEmployee):
    def __init__(self, name, id, salary, hours_per_week, mentor):
        super().__init__(name, id, salary, hours_per_week)
        self.mentor = mentor

    def display_details(self):
        super().display_details()
        print(f"Mentor: {self.mentor}")

# Creating objects for each class and displaying details
employee = Employee("John Doe", "E001", 50000)
full_time_employee = FullTimeEmployee("Alice Smith", "E002", 70000, "Health, Dental, 401k")
part_time_employee = PartTimeEmployee("Bob Brown", "E003", 30000, 20)
manager = Manager("Carol White", "E004", 90000, "Health, Dental, 401k", 10)
developer = Developer("Dave Black", "E005", 80000, "Health, Dental, 401k", ["Python", "Java", "C++"])
intern = Intern("Eve Green", "E006", 20000, 15, "John Doe")

print("Employee Details:")
employee.display_details()
print("\nFull-Time Employee Details:")
full_time_employee.display_details()
print("\nPart-Time Employee Details:")
part_time_employee.display_details()
print("\nManager Details:")
manager.display_details()
print("\nDeveloper Details:")
developer.display_details()
print("\nIntern Details:")
intern.display_details()
```

```python
class Employee:
  def __init__(self, name, id, salary):
    self.name = name
    self.id = id
    self.__salary = salary

  def get_salary(self):
    return self.__salary

  def display_details(self):
    print(f"Name of Employee:{self.name}")
    print(f"ID of Employee: {self.id}")
    print(f"Salary of Employee: {self.__salary}")

# Intermediate Classes

class FulltimeEmployee(Employee):
  def __init__(self, name, id, salary, benefits):
    super().__init__(name, id, salary)
    self.__benefits = benefits

  def get_benefits(self):
    return self.__benefits

  def display_details(self):
    super().display_details()
    print(f"The salary benefits of Employee: {self.__benefits}")

class PartTimeEmployee(Employee):  # Single inheritance (inherits from Employee)
    def __init__(self, name, id, salary, hours_per_week):
        super().__init__(name, id, salary) #no need to use self
        self.hours_per_week = hours_per_week

    def display_details(self):
        super().display_details()
        print(f"Hours per week: {self.hours_per_week}")

#Derived class
class Manager(FulltimeEmployee):
  def __init__(self, name, id, salary, benefits, team_size):
    FulltimeEmployee.__init__(self,name, id, salary, benefits) #always use self
    self.__team_size = team_size

  def get_team(self):
    self.__team_size

  def display_details(self):
    super().display_details()
    print(f"The Team_size of the Team: {self.__team_size}")

class Developer(FulltimeEmployee): # Multilevel inheritance (inherits from FullTimeEmployee which inherits from Employee)
  def __init__(self, name, id, salary, benefits, programming_language):
    super().__init__(name, id, salary, benefits)
    self.__programming_language = programming_language

  def get_pro_lan(self):
    return self.__programmming_language

  def dipslay_details(self):
    super().display_details()
    print(f"The programming_language: {self.__programming_language}")

class Intern(PartTimeEmployee):
  def __init__(self, name, id,salary, hours_per_week, mentor):
    PartTimeEmployee.__init__(self,name, id, salary, hours_per_week)
    self.__mentor = mentor

  def get_mentor(self):
    return self.__mentor

  def display_details(self):
     super().display_details()
     print(f"The Mentor is : {self.__mentor}")

# Creating objects for each class and displaying details
employee = Employee("John Doe", "E001", 50000)  # Base class instance
full_time_employee = FullTimeEmployee("Alice Smith", "E002", 70000, "Health, Dental, 401k")  # Single inheritance
part_time_employee = PartTimeEmployee("Bob Brown", "E003", 30000, 20)  # Single inheritance
manager = Manager("Carol White", "E004", 90000, "Health, Dental, 401k", 10)  # Multilevel inheritance
developer = Developer("Dave Black", "E005", 80000, "Health, Dental, 401k", ["Python", "Java", "C++"])  # Multilevel inheritance
intern = Intern("Eve Green", "E006", 20000, 15, "John Doe")  # Multilevel inheritance


#Print all
print("Employee Details:")
employee.display_details()
print("\nFull-Time Employee Details:")
full_time_employee.display_details()
print("\nPart-Time Employee Details:")
part_time_employee.display_details()
print("\nManager Details:")
manager.display_details()
print("\nDeveloper Details:")
developer.display_details()
print("\nIntern Details:")
intern.display_details()




```
