# **Explain Polymorphism with example and write python code. **

Polymorphism is an important concept in OOPS.It is the ability of the object of different classes to take more than one form using the same method or same function call.
It provides code modularity
Reusability


```python
class Animal:
  def make_sound(self):
    print("Generic sound of animals")

  def __init__(self,name,age):
    self.__name = name
    self.age = age

  def get_name(self):
    return self.__name
    print(f'The animal {self.get_name()} makes a sound')

  def get_age(self):
    return self.age
    print(f'The animal {self.get_name()} makes a good sound')

class Dog(Animal):
  def make_sound(self):
    print("woof")

class Cat(Animal):
  def make_sound(self):
    print("Meow")

#creates Polymorphism
def speak(animal):
  animal.make_sound() #here its implemented same function call

#creates object or instances for class Dog and Cat
dog = Dog("Dash",4)
cat = Cat("Kit",3)

speak(dog)
speak(cat)



```

# **3. Explain Single Level Inheritance with python code**

Inheritance is the concept in OOPS where a subclass inherits properties from Base class or (super class or parent class)

**Single level inheritance**

---

Its the inheritance in which there is only one base class/Derived class.
That is a derived class inherites from a single base class only.

```python
#Base class or Parent class
class Animal:
  def make_sound(self):
    return "roar"

#Derived class or child class or sub class
class Dog(Animal):  #Inherits properties attributes and methods from superclass Animal
  def bark(self):
    return "woof"
#Create instance for subclass Dog
dog = Dog()

#print(dog.make_sound())
print(dog.bark())
```

# **4. Explain Multiple inheritance with python code.

---


Multiple inheritance is the inheritance in which a subclass inherits properties or attributes from multiple parent classes or superclasses.

```python
class Father:
  def skills(self):
    print("Fathers skill-He is a good surgon")

class Mother:
  def skills(self):
    print("Mothers skills- She is a very good scientist")

class Child(Father, Mother):
  def skills(self):
    super().skills()  #use the concept of MRO-(Method Resolution Order)
    print("Child skills-she sings and dance")

child = Child()
child.skills()
```

Here MRO is the order in which the subclass determines the sequence in which the object or the function hs to be outputted or displayed since they have got same function name for both parents and child.It usually follows c3 Lineraisation and precidence from left to right.

# **5. Explain Muti-level inheritance with python code **

Multi level inheritance is the inheritance in which a derived class inherites from a derived  class which inherites from a base class.

```python
class Animal:
  def speak(self):
    print("roar")

class Dog(Animal):
  def bark(self):
    print("woof")

class Cat(Dog):
  def cry(self):
    print("meow")


dog = Dog()
cat = Cat()

dog.speak()
dog.bark()

cat.speak()
cat.bark()
cat.cry()
```

# **Write a program of factorial in python **

```python
def fact(n):
  if n==0:
    return 1
  else:
    return n*fact(n-1)

print(fact(5))
print(fact(4))
```

# **10. How many pillars of Oops we have in Python?**

There are 5 pillars for OOPS(Object oriented Programing).

They are
1: objects- which are instances of a class

2: classes- Its the blueprint for creating an object

3:Inheritance- Its the property by which a sub class or child class inherits all properties like attributes and methods from a parent class or base class.

4:Polymorphism - Its is the ability of an object to take more forms from different classes using the same function call.The allows us to modify the code when needed.

5:Encapsulation - It is the bundling of data and methods into a single unit using class.This property allows data hiding so that code wont be misused.

All this properties allow oops to provide better usage of code,abstraction,reusability and flexibility.
