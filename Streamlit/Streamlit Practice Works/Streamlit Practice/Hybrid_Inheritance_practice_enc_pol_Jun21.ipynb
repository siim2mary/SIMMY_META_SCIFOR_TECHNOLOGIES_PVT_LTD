```python
# Base class Univ with encapsulation
class Univ:
    def __init__(self, name, age, add):
        self.__name = name
        self.age = age
        self.__add = add

    def get_name(self):
        return self.__name

    def get_add(self):
        return self.__add

    def get_role(self):
        print(f'The name of person in university is {self.__name} and Age: {self.age} and Address: {self.__add}')
        return "Univ"

# Derived class Stud
class Stud(Univ):
    def __init__(self, name, age, add, std_id, dept):
        super().__init__(name, age, add)
        self.__dept = dept
        self.__std_id = std_id

    def get_dept(self):
        return self.__dept

    def get_std_id(self):
        return self.__std_id

    def display_details(self):
        print(f'The name of student is Name: {self.get_name()}, Age: {self.age}, Address: {self.get_add()}, Student ID: {self.get_std_id()}, Department: {self.get_dept()}')

    def get_role(self):
        return "Stud"

# Derived class Teacher
class Teacher(Univ):
    def __init__(self, name, age, add, subject):
        super().__init__(name, age, add)
        self.__subject = subject

    def get_subject(self):
        return self.__subject

    def display_details(self):
        print(f'The name of teacher is Name: {self.get_name()}, Age: {self.age}, Address: {self.get_add()}, Subject: {self.get_subject()}')

    def get_role(self):
        return "Teacher"

# Additional class Employee
class Employee:
    def __init__(self, emp_id, salary):
        self.__emp_id = emp_id
        self.__salary = salary

    def get_emp_id(self):
        return self.__emp_id

    def get_salary(self):
        return self.__salary

# Derived class Staff with multiple inheritance
class Staff(Univ, Employee):
    def __init__(self, name, age, add, emp_id, salary, dept):
        Univ.__init__(self, name, age, add)
        Employee.__init__(self, emp_id, salary)
        self.__dept = dept

    def get_dept(self):
        return self.__dept

    def display_details(self):
        print(f'The name of staff is Name: {self.get_name()}, Age: {self.age}, Address: {self.get_add()}, Employee ID: {self.get_emp_id()}, Salary: {self.get_salary()}, Department: {self.get_dept()}')

    def get_role(self):
        return "Staff"

# Hybrid inheritance
class Office(Teacher, Employee):
    def __init__(self, name, age, add, dept, subject, emp_id, salary, standard):
        Teacher.__init__(self, name, age, add, subject)
        Employee.__init__(self, emp_id, salary)
        self.__standard = standard
        self.__dept = dept

    def get_standard(self):
        return self.__standard

    def get_dept(self):
        return self.__dept

    def display_details(self):
        print(f'The name of office staff is Name: {self.get_name()}, Age: {self.age}, Address: {self.get_add()}, Subject: {self.get_subject()}, Employee ID: {self.get_emp_id()}, Salary: {self.get_salary()}, Standard: {self.get_standard()}, Department: {self.get_dept()}')

    def get_role(self):
        return "Office"

# Create instances
st = Stud("Harry", 23, "Park avenue", "STD-1234", "Physics")
tea = Teacher("Preethy", 39, "789 Boulevard", "Mechanics")
staf = Staff("Rony", 35, "Rosario street", "EMP-3333", 350000, "Physics")
off = Office("Rocky", 40, "456 Avenue", "Commerce", "Economics", "EMP-4444", 400000, 3)

# Test the methods
st.display_details()
tea.display_details()
staf.display_details()
off.display_details()

# Demonstrating polymorphism
def print_role(person):
    print(f"{person.get_name()} is a {person.get_role()}.")

print_role(st)
print_role(tea)
print_role(staf)
print_role(off)
```
